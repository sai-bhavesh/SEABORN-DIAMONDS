```python
import seaborn as sns
import matplotlib.pyplot as plt
var=sns.load_dataset('diamonds')
var
```

# 1.WHAT IS THE DISTRIBUTION OF DIAMOND PRICES?

```python
sns.histplot(x='price',data=var,kde=True,color='teal')
plt.legend()
plt.title('WHAT IS THE DISTRIBUTION OF DIAMOND PRICES?')

plt.show()
```

# 2.HOW DOES DIAMOND PRICE VARY BY CUT?

```python
plt.figure(figsize=(5,20))
plt.title('PRICE VARIATION BY CUT')
sns.boxplot(x='cut',y='price',data=var,color='turquoise')
plt.show()
```

# 3. WHAT IS THE RELATIONSHIP BETWEEN CARAT AND PRICE?

```python
plt.figure(figsize=(20,20))
plt.title('RELATIONSHIP BETWEEN CARAT AND PRICE')
sns.scatterplot(x='carat',y='price',data=var,hue='cut')
plt.show()
```

# 4. HOW DOES THE PRICE VARY ACROSS DIFFERENT LEVELS OF DIAMOND CLARITY?

```python
plt.title('PRICE VARIATION ACROSS DIFFERENT LEVELS OF DIAMOND CLARITY?')
sns.boxplot(x='clarity',y='price',data=var,color='mediumpurple')
plt.show()
```

# 5.WHAT’S THE COUNT OF DIAMONDS AVAILABLE IN EACH CUT CATEGORY?

```python
plt.title('COUNT OF DIAMONDS AVAILABLE IN EACH CUT CATEGORY')
sns.histplot(x='cut',data=var,color='blueviolet')
plt.show()
```

```python
# just another way...but should always use count plot for categorical values

plt.title('COUNT OF DIAMONDS AVAILABLE IN EACH CUT CATEGORY')
sns.countplot(x='cut',data=var,color='blueviolet')
plt.show()
```

# 6.HOW DOES THE NUMBER OF DIAMONDS VARY BY CUT, FURTHER GROUPED BY COLOR?

```python
sns.countplot(x='cut',data=var,hue='color',color='turquoise')
plt.show()
```

# 7.WHAT FEATURES IN THE DATASET ARE STRONGLY CORRELATED WITH DIAMOND PRICE?

```python
numeric=var.select_dtypes(include=['float64','int64'])
corr=numeric.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f')
plt.show()
```

# 8.HOW DOES DIAMOND PRICE VARY WITH CARAT ACROSS DIFFERENT CLARITY LEVELS?

```python
plt.figure(figsize=(20,20))
sns.scatterplot(x='carat',y='price',data=var,hue='cut',color='mediumseagreen')
plt.show()
```

# 9.HOW DOES DIAMOND PRICE VARY BY BOTH CUT AND COLOR?

```python
sns.barplot(x='cut',y='price',data=var,hue='color')
plt.show()
```

# 10. HOW DOES THE DISTRIBUTION OF DIAMOND PRICE VARY ACROSS CLARITY LEVELS?

```python
sns.violinplot(x='clarity',y='price',data=var,palette='cool')
plt.show()
```

# 💎 Exploratory Data Analysis (EDA) on Diamonds Dataset using Seaborn

## 📌 Objective:
Perform in-depth EDA on the Seaborn Diamonds dataset to explore how different features affect the price of diamonds using visualizations.

## 📊 Dataset Overview:
- **Total Records**: 53,940
- **Features**:
  - *Categorical*: `cut`, `color`, `clarity`
  - *Numerical*: `carat`, `depth`, `table`, `price`, `x`, `y`, `z`

## 🔍 EDA Summary:

1. **Distribution of Carat**  
   → Right-skewed; most diamonds are less than 1 carat.

2. **Price Distribution by Cut**  
   → `Ideal` cut is most common but not the most expensive.

3. **Carat vs Price**  
   → Strong positive correlation; higher carat = higher price.

4. **Price vs Clarity**  
   → Median price varies across clarity; `IF` tends to be pricier.

5. **Count of Diamonds by Cut**  
   → `Ideal` dominates; `Fair` is least common.

6. **Count by Cut and Color**  
   → Some color categories are more frequent in certain cut types.

7. **Correlation Heatmap**  
   → `carat` has the highest correlation with `price`; `x`, `y`, `z` also correlate.

8. **Carat vs Price (Grouped by Clarity)**  
   → Shows how clarity affects pricing trends across carats.

9. **Avg. Price by Cut and Color**  
   → Clear differences in average price between color groups under each cut.

10. **Price Distribution by Clarity (Violinplot)**  
    → Insight into spread, density, and outliers by clarity.

## ✅ Tools Used:
- Seaborn
- Matplotlib
- Pandas
- Jupyter Notebook

## 🧠 Final Insights:
- **Carat** is the most influential feature on price.
- Categorical variables like **cut** and **clarity** show clear patterns when visualized.
- Using `hue` adds major insight into subgroup behavior.
- Violin and scatter plots help spot patterns and outliers effectively.
